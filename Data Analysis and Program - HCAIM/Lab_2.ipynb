Task 5 - There are <sup/>7</sup>C<sub/>2</sub> possible bivariate plots possible for the given dataset. You need to create any 6 bivariate plots of your choice and interpret the results. 

Interpretation of results are at the end of graphs to make it a cleaner notebook style.

```python
# Scatter plot of total bill and tip
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('Scatterplot of Total Bill and Tip')
plt.show()
```

```python
# Boxplot of total bill and time
sns.boxplot(x='time',y='total_bill',data=df)
plt.title('Boxplot of Total Bill by Time')
plt.show()
```

```python
# Barplot of day and tip
sns.barplot(x='day',y='tip',data=df)
plt.title('Barplot of Day and Tip')
plt.show()
```

```python
# Pointplot of sex and total bill
sns.pointplot(x='sex',y='total_bill',data=df)
plt.title('Pointplot of Sex and Total Bill')
plt.show()
```

```python
# Violinplot of smoker and total bill
sns.violinplot(x='smoker',y='total_bill',data=df)
plt.title('Violinplot of Smoker and Total Bill')
plt.show()
```

```python
# Swarmplot of smoker and total bill
sns.swarmplot(x='smoker',y='total_bill',data=df)
plt.title('Swarmplot of Smoker and Total Bill')
plt.show()
```
